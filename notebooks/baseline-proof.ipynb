# Introduction

This notebook establishes a baseline for assessing whether a Research Agent can produce an acceptable `DirectProof` from an input document.
The baseline is intentionally narrow: it is meant to show what mistakes the Research Agent makes before we add fuller axiomatization, richer reasoning profiles, or additional runtime affordances.

Evaluation is delegated to a reusable proof evaluation harness rather than being defined ad hoc in this notebook.
That keeps the notebook focused on examples, observed failure modes, and conclusions grounded in actual harness output.
